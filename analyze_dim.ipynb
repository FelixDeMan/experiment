```python
import pandas as pd


df = pd.read_csv('fixed_scores_og_amount.csv')
```

```python
df.head()
```

```python
df['model_no_dim'] = "UNet_lr"+df['UNet_lr'] + 'TE_lr' + df['TE_lr']
df['model_no_dim']
```

```python
idx = df.groupby('model_no_dim')['normalized_score'].idxmax()

# Select the rows corresponding to the obtained indices
best_entries = df.loc[idx]
```

```python
print(best_entries.shape)
best_entries.head()
```

```python
dim_counts = best_entries['dim'].value_counts()
dim_counts
```

```python
best_entries_sorted = best_entries.sort_values('normalized_score', ascending=False)
best_entries_sorted
```

```python
best_entries.to_csv('lora_per_model.csv')
```
