```python


def fix_pic_amount(df):
  
    df['ocr_scores'] = df['ocr_scores'].apply(lambda x: x.split(',')[:48])
    df['confidence_scores_processed'] = df['confidence_scores_processed'].apply(lambda x: x.split(',')[:48])

    return df
```

```python
df_fixed = fix_pic_amount(df)
df_fixed
```

```python
df_fixed['amount_pics'] = df_fixed['ocr_scores'].apply(len)
unique_counts = df_fixed['amount_pics'].value_counts()
unique_counts
```

```python
from statistics import mean
df_fixed['confidence_scores_processed'].apply(lambda x:print( eval(x)))
df_fixed['confidence_scores_corrected'] = df_fixed['confidence_scores_processed'].apply(lambda x:mean(eval(x)))
df_fixed['confidence_scores_corrected']
```

```python
min_value = df_fixed['confidence_scores_corrected'].min()
max_value = df_fixed['confidence_scores_corrected'].max()

df_fixed['normalized_yolo'] = (df_fixed['confidence_scores_corrected'] - min_value) / (max_value - min_value)
```
